```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Hiển thị biểu đồ ngay trong notebook
%matplotlib inline

# Cấu hình thẩm mỹ cho biểu đồ
sns.set_style('whitegrid')
palette = sns.color_palette("viridis")

# Cấu hình hiển thị dataframe (hiển thị hết các cột)
pd.set_option('display.max_columns', None)
```

```python
# Tải dữ liệu
df = pd.read_csv('../data/prosperLoanData.csv')

# Kiểm tra 5 dòng đầu tiên
print("Dữ liệu 5 dòng đầu:")
display(df.head())

# Kiểm tra thông tin các cột và kiểu dữ liệu
print("\nThông tin bộ dữ liệu:")
df.info()

# Kiểm tra các giá trị thống kê cơ bản
print("\nThống kê mô tả:")
display(df.describe())
```

```python
# 1. Chọn ra các cột quan trọng nhất để phân tích sâu
cols = ['LoanOriginalAmount', 'BorrowerRate', 'Term', 'LoanStatus', 
        'ProsperRating (Alpha)', 'ListingCategory (numeric)', 'EmploymentStatus',
        'StatedMonthlyIncome', 'DebtToIncomeRatio', 'LoanOriginationDate']
loans = df[cols].copy()

# 2. Chuyển đổi định dạng ngày tháng
loans['LoanOriginationDate'] = pd.to_datetime(loans['LoanOriginationDate'])

# 3. Lọc bỏ các dòng không có ProsperRating (thường là các khoản vay cũ trước 2009)
loans = loans.dropna(subset=['ProsperRating (Alpha)'])

# 4. Chuyển mã số Category thành tên thực tế
categories = {0: 'Not Available', 1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 
              7: 'Other', 15: 'Medical/Dental', 18: 'Taxes', 19: 'Vacation'}
# (Bạn có thể thêm các mã khác nếu cần)
loans['ListingCategory'] = loans['ListingCategory (numeric)'].map(lambda x: categories.get(x, 'Other'))

# 5. Sắp xếp thứ tự các mức xếp hạng tín dụng (AA là tốt nhất, HR là rủi ro nhất)
rating_order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR']
loans['ProsperRating (Alpha)'] = pd.Categorical(loans['ProsperRating (Alpha)'], 
                                                categories=rating_order, ordered=True)

print("Làm sạch dữ liệu hoàn tất!")
```

```python
plt.figure(figsize=(16, 10))

# Biểu đồ 1: Phân phối số tiền vay (Không đổi vì không dùng palette)
plt.subplot(2, 2, 1)
sns.histplot(loans['LoanOriginalAmount'], bins=30, kde=True)
plt.title('Phân phối số tiền vay gốc')

# Biểu đồ 2: Trạng thái khoản vay (Cập nhật hue)
plt.subplot(2, 2, 2)
sns.countplot(data=loans, y='LoanStatus', 
              hue='LoanStatus', palette='viridis', legend=False,
              order=loans['LoanStatus'].value_counts().index)
plt.title('Thống kê trạng thái các khoản vay')

# Biểu đồ 3: Phân phối lãi suất (Không đổi)
plt.subplot(2, 2, 3)
sns.violinplot(data=loans, x='BorrowerRate', color='skyblue')
plt.title('Mật độ phân bổ lãi suất của người vay')

# Biểu đồ 4: Xếp hạng tín dụng (Cập nhật hue)
plt.subplot(2, 2, 4)
sns.countplot(data=loans, x='ProsperRating (Alpha)', 
              hue='ProsperRating (Alpha)', palette='RdYlGn_r', legend=False)
plt.title('Số lượng khoản vay theo Xếp hạng Prosper')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 6))

# 1. Xếp hạng tín dụng vs Lãi suất (Cập nhật hue)
plt.subplot(1, 2, 1)
sns.boxplot(data=loans, x='ProsperRating (Alpha)', y='BorrowerRate', 
            hue='ProsperRating (Alpha)', palette='RdYlGn_r', legend=False)
plt.title('Mối quan hệ: Xếp hạng tín dụng vs Lãi suất')

# 2. Thu nhập hàng tháng vs Số tiền vay (Không đổi)
plt.subplot(1, 2, 2)
income_filter = loans[loans['StatedMonthlyIncome'] < 20000]
sns.regplot(data=income_filter.sample(5000), x='StatedMonthlyIncome', y='LoanOriginalAmount', 
            scatter_kws={'alpha':0.1}, line_kws={'color':'red'})
plt.title('Tương quan: Thu nhập hàng tháng vs Số tiền vay')

plt.show()
```

```python
plt.figure(figsize=(10, 8))
# Chỉ lấy các cột kiểu số
numeric_cols = loans.select_dtypes(include=[np.number])
corr = numeric_cols.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Ma trận tương quan giữa các biến số')
plt.show()
```
